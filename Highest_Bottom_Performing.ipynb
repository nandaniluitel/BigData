```python
import pandas as pd
school_df = pd.read_csv("/content/drive/MyDrive/pandas/schools_complete.csv")
```

```python

student_df = pd.read_csv("/content/drive/MyDrive/pandas/students_complete.csv")
complete_data = pd.merge(school_df,student_df,how="left",on=["school_name","school_name"])
```

```python
complete_data.head(10)
```

```python
# prompt: Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.

# Calculate the overall passing rate
complete_data['% Overall Passing'] = (complete_data['reading_score'] >= 70) & (complete_data['math_score'] >= 70)

# Group data by school name and calculate the mean of overall passing rate
overall_passing_per_school = complete_data.groupby('school_name')['% Overall Passing'].mean() * 100

# Sort schools by % Overall Passing in descending order and display the top 5
top_5_schools = overall_passing_per_school.sort_values(ascending=False).head(5)
top_5_schools
```

```python
# prompt: # Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.

# Sort schools by % Overall Passing in ascending order and display the top 5
bottom_5_schools = overall_passing_per_school.sort_values(ascending=True).head(5)
bottom_5_schools
```
