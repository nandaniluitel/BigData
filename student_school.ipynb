```python
import pandas as pd
school_df = pd.read_csv("/content/drive/MyDrive/pandas/schools_complete.csv")
```

```python

student_df = pd.read_csv("/content/drive/MyDrive/pandas/students_complete.csv")
complete_data = pd.merge(school_df,student_df,how="left",on=["school_name","school_name"])
```

```python
complete_data.head(10)
```

```python
# prompt: select the school type

school_types = school_df.set_index(["school_name"])["type"]
school_types
```

```python
# prompt: calculate the total number of students per school

per_school_counts = complete_data["school_name"].value_counts()
per_school_counts
```

```python
# prompt: calculate  the total school budget for each school and per capita spending for each school
# Convert budget to numeric if necessary
complete_data["budget"] = pd.to_numeric(complete_data["budget"], errors="coerce")
# Calculate the total budget for each school
per_school_budget = complete_data.groupby("school_name")["budget"].mean()

# Calculate the total number of students in each school
per_school_counts = complete_data.groupby("school_name")["size"].mean()

# Calculate per capita spending for each school
per_school_capita = per_school_budget / per_school_counts

print("Total School Budget (Per School):")
print(per_school_budget)

print("\nPer Capita Spending (Per School):")
print(per_school_capita)
```

```python
# prompt: calculate the average test scores per school

# Calculate the average math score per school
per_school_math = complete_data.groupby(["school_name"])["math_score"].mean()

# Calculate the average reading score per school
per_school_reading = complete_data.groupby(["school_name"])["reading_score"].mean()

# Calculate the average test scores per school
per_school_passing_math = complete_data[complete_data["math_score"] >= 70].groupby(["school_name"])["Student ID"].count()/per_school_counts * 100
per_school_passing_reading = complete_data[complete_data["reading_score"] >= 70].groupby(["school_name"])["Student ID"].count()/per_school_counts * 100
overall_passing_rate = (per_school_passing_math + per_school_passing_reading) / 2

# Print the results
print("\nAverage Math Score Per School:")
print(per_school_math)

print("\nAverage Reading Score Per School:")
print(per_school_reading)

print("\nOverall Passing Rate Per School:")
overall_passing_rate
```

```python
# prompt: calculate  the number of school with math scores of 70 or higher

passing_math_by_school = complete_data[complete_data["math_score"] >= 70].groupby("school_name")["Student ID"].count()
school_passing_math_count = passing_math_by_school.count()
print(f"Number of schools with math scores of 70 or higher: {school_passing_math_count}")
```

```python
# prompt: calculate  the number of school with reading scores of 70 or higher

passing_reading_by_school = complete_data[complete_data["reading_score"] >= 70].groupby("school_name")["Student ID"].count()
school_passing_reading_count = passing_reading_by_school.count()
print(f"Number of schools with reading scores of 70 or higher: {school_passing_reading_count}")
```

```python
# prompt: calculate the schools that passed both math and reading with scores of 70 or higher

# Calculate the number of schools where students passed both math and reading with scores of 70 or higher
passing_math_reading = complete_data[(complete_data["math_score"] >= 70) & (complete_data["reading_score"] >= 70)]
schools_passing_both = passing_math_reading.groupby("school_name")["Student ID"].count()
num_schools_passing_both = schools_passing_both.count()

print(f"Number of schools where students passed both math and reading: {num_schools_passing_both}")
```

```python
# prompt: Calculate the passing rates per school

per_school_passing_math = complete_data[complete_data["math_score"] >= 70].groupby(["school_name"]).count()["student_name"]
per_school_passing_reading = complete_data[complete_data["reading_score"] >= 70].groupby(["school_name"]).count()["student_name"]

per_school_passing_math = per_school_passing_math / per_school_counts * 100
per_school_passing_reading = per_school_passing_reading / per_school_counts * 100

overall_passing_rate = (per_school_passing_math + per_school_passing_reading) / 2

overall_passing_rate
```

```python
per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "Total Budget": per_school_budget,
    "Per Capita Spending": per_school_capita,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "Overall Passing Rate Per School":overall_passing_rate,
    "Math scores of 70 or higher": school_passing_math_count,
    "Reading scores of 70 or higher": school_passing_reading_count,
    "Both math and reading with scores of 70 or higher": num_schools_passing_both,
    "% Overall Passing per School": overall_passing_rate
})
# Reset the index to make 'school_name' a column
per_school_summary = per_school_summary.reset_index()

# Display the updated DataFrame
print(per_school_summary)
```


# Scores by School Spending

```python
# prompt: # Establish the bins

# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]


```

```python
# prompt: # Create a copy of the school summary since it has the "Per Student Budget"

# Create a copy of the school summary DataFrame
school_spending_df = per_school_summary.copy()

```

```python
# prompt: # Use `pd.cut` to categorize spending based on the bins.

# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Capita Spending"], spending_bins, labels=labels, right=False)
school_spending_df
```

```python
# prompt: #  Calculate averages for the desired columns.

# Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Math scores of 70 or higher"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Reading scores of 70 or higher"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing per School"].mean()

# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

# Display results
spending_summary
```
